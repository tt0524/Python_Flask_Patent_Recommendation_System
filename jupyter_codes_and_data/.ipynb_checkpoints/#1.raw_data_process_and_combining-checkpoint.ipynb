```python
df = pd.read_csv("patent_sample.csv").drop(['Unnamed: 0'], axis=1)
dfk = df.copy()
```

```python
df_combine = dfk[['id','date','abstract','title','kind','num_claims','A','B','C','D','E','F','G','H','Y']]
df_combine.head()
```

```python
inv = pd.read_csv("Data/inv_cleaned.csv").drop(['Unnamed: 0'], axis=1)
law = pd.read_csv("Data/law_cleaned.csv").drop(['Unnamed: 0'], axis=1)
asi = pd.read_csv("Data/asi_cleaned.csv").drop(['Unnamed: 0'], axis=1)
```

```python

import numpy as np
inv_combine = inv.copy()
inv_combine['inventor_name'] = inv_combine['name_first']+' '+inv_combine['name_last']
inv_combine = inv_combine.drop(columns=['name_first', 'name_last','inventor_id'])
inv_combine = inv_combine.groupby(['patent_id']).agg(lambda x: tuple(x)).applymap(list).reset_index()
inv_combine.head()
```

```python
law_combine = law.copy()
law_combine = law_combine.fillna('')
law_combine['lawyer_name'] = law_combine['name_first']+' '+law_combine['name_last']+','+law_combine['organization']
law_combine = law_combine.drop(columns=['name_first', 'name_last','lawyer_id','organization'])
law_combine = law_combine.groupby(['patent_id']).agg(lambda x: tuple(x)).applymap(list).reset_index()
law_combine.head()
```

```python
asi_combine = asi.copy()
asi_combine = asi_combine.fillna('')
asi_combine['assignee_name'] = asi_combine['name_first']+' '+asi_combine['name_last']+','+asi_combine['organization']
asi_combine = asi_combine.drop(columns=['name_first', 'name_last','assignee_id','organization','type'])
asi_combine = asi_combine.groupby(['patent_id']).agg(lambda x: tuple(x)).applymap(list).reset_index()
asi_combine.head()
```

```python
names = pd.merge(inv_combine, law_combine, left_on='patent_id', right_on='patent_id',how="outer")
names = pd.merge(names, asi_combine, left_on='patent_id', right_on='patent_id',how="outer")
names.fillna('unknown', inplace=True)
names.head()
```

```python
df_final = pd.merge(df_combine, names, left_on='id', right_on='patent_id', how="outer")
df_final = df_final.drop(columns=['patent_id'])
df_final.to_csv('Data/all_combined.csv')
```
