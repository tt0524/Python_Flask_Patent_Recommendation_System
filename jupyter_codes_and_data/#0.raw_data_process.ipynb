```python
cpc_current = cpc_current[['patent_id','section_id']]
```

```python
patent_join = patent_join.drop(columns=['kind','filename','withdrawn','patent_id'])
```

```python
patent_section = patent_join[['id','section_id']]
```

```python
patent_section['dummy'] = 1
```

```python
patent_section_pivot = patent_section.pivot(index='id', columns='section_id',values='dummy')
```
